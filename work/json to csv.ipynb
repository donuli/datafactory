```python
import json
import numpy as np
```

```python
filename = 'instances_default.json'

with open(filename) as json_file:
    json_data = json.load(json_file)

categories = json_data["categories"]
images = json_data["images"]
annotations = json_data["annotations"]
```

```python
def annotation():
    category_id={}
    image_id={}
    annotation_box=[]
    
    for category in categories:
        category_id[category["id"]]=category["name"]
    
    for image in images:
        image_id[image["id"]]=image["file_name"]
    
    for annotation in annotations:
        category_num = annotation["category_id"]
        image_num = annotation["image_id"]
        bbox = annotation["bbox"]
        attributes = annotation["attributes"]
        status = attributes["status"]
        annotation_box.append({"label_name":category_id[category_num], "image_name" : image_id[image_num] , "bbox" : bbox, "status" : status})
    print(type(bbox))
    return annotation_box
```

```python
annotation()
```

```python
label_info=[]
for i in annotation():
    box = i["bbox"]
    label_info.append({"image": i["image_name"], "x":box[0], "y":box[1], "width":box[2], "hight":box[3], "label":i["label_name"]+"_"+i["status"]})
```

```python
print(label_info)
```

```python
import csv
```

```python
file = open("dataset.csv",mode='w')
writer = csv.writer(file)
writer.writerow(["image","x","y","width","hight","label"])
for i in label_info:
    writer.writerow(list(i.values()))
```
